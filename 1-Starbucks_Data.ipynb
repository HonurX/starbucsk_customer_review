```python
y = df["20. Will you continue buying at Starbucks?"].values
x = df.drop("20. Will you continue buying at Starbucks?" , axis =1 ).values
```

```python
y.shape
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split

x_train , x_test , y_train , y_test  = train_test_split(x, y , random_state=34 , test_size=0.1)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()

x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
x_train.shape
```

```python
y_test.shape
```

```python
y_train.shape
```

```python
import tensorflow as tf
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense , Activation , Dropout
from sklearn.metrics import classification_report , confusion_matrix
```

```python
modelTensor = Sequential()
```

```python
modelTensor.add(Dense(19, activation="relu"))
modelTensor.add(Dense(10, activation="relu"))
modelTensor.add(Dense(10, activation="relu"))
modelTensor.add(Dense(10, activation="relu"))
modelTensor.add(Dense(2,activation="relu"))

modelTensor.add(Dense(1,activation="sigmoid"))

modelTensor.compile(optimizer="adam" , loss = "binary_crossentropy" , metrics=(["accuracy"]) )

                    
```

```python
modelTensor.fit(x=x_train , y=y_train  , verbose=1 , epochs=150, validation_data=(x_test , y_test))
```

```python
modelLoss = pd.DataFrame(modelTensor.history.history)
```

```python
modelLoss.plot()
```

```python
yPred= modelTensor.predict(x_test)
```

```python
yPred
```

```python
predicted_classes = (yPred > 0.6).astype(int)
```

```python
predicted_classes
```

```python
print(classification_report(y_test, predicted_classes))
```

```python
print(confusion_matrix(y_test , predicted_classes))
```
